# Conceito do Stash

- O stash é uma funcionalidade do Git que permite guardar temporariamente alterações não finalizadas na working tree. Em situações em que é necessário mudar de contexto ou ajustar outra parte do projeto sem perder as modificações que ainda não estão prontas para serem comitadas, o stash age como uma área de armazenamento provisório. Dessa forma, evita-se a necessidade de realizar parciais que podem poluir o histórico do repositório.

[-] Ao usar o comando `git stash`, o Git captura as alterações atuais do diretório e do stage, salvando-as em um apilha interna. A working tree é, então, restaurada para o último commit, proporcionando um ambiente limpo paras novas alterações. Quando for o momento de retomar o trabalho anterios, 0 comando `git stash pop` resgata as modificações armazenadas e as reaplica na working tree, removendo-as das pilha de stashes.


### Armazena as alterações atuais
`git stash`

### Verifica a lista de stashes salvos
`git stash list`

### Reaplica as alterações e remove o stash da pilha
`git stash pop`

# Funcionamento do HEAD

 O  HEAD é um ponteiro especial que indica a posição atual dentro do histórico do repositótio. Em essência, ele aponta para o commit que está atualmente em uso, funcionando como uma referência dinâmica para a branch ativa. Internamente, o Git guarda essa informação num arquivo simples, permitindo que as operações de commit e checkout saibam exatamente qual é o estado atual da árvores de código.

 Para inspecionar o conteúdo do HEAD, pode-se utilizar o comando `cat .git/HEAD`, que geralmente retorna algo como `ref: refs/heads/main`, sinalizando que o HEAD aponta para a branch "main". Essa abordagem permite que o Git mantenha um fluco consistente de mudanças, pois todas as nova alterações serão registradas a partir desse ponto de referência.

 # A Relação entre HEAD e Branches

 Ao alternar entre branches, o Git atualiza o HEAD para apontar para a nova referência. Por exemplot, ao executar `git checkout (nome da branch)`, o HEAD muda de forma imediata, direcionando as operações subsequentes de commit e merge para o escopo desta branch. Essa dinâmica é fundamental para garantir que as alterações sejam isoladas, facilitando a experimentação sem interferir com outros desenvolvimentos paralelos.

 A utilização do HEAD confere maior segurança e clareza à hora de integrar mudanças. Quando o HEAD está corretamente apontado, é mais simples identificar o ponto de partida para novas funcionalidades ou correções, permitindo uma jornada ordenada e rastreável do projeto.

```python
# Cria um dataframe com os países da América do Sul
america_sul = df.query('Região == "América do Sul"').sort_values(by='Total', ascending=True)

# Criando uma lista de cores para destacar o Brasil
cores = []
for pais in america_sul.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

# Criando a figura
fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul.index, america_sul['Total'], color = cores)
ax.set_xlabel("Número de imigrantes", fontsize = 14)
ax.set_ylabel("")
ax.yaxis.set_tick_params(labelsize = 12)
ax.xaxis.set_tick_params(labelsize = 12)

#Loop para adcionar o valor máximo de cada barra
for i,v in enumerate(america_sul['Total']):
    print(i,v)
    ax.text(v + 20, i, str(v), color = 'k', fontsize = 10, ha ='left', va ='center')

import os
nome_da_pasta = "imagens"

if not os.path.isdir(nome_da_pasta):
    os.makedirs(nome_da_pasta)
    print(f"Pasta {nome_da_pasta} criada.")
else:
    print(f"Pasta {nome_da_pasta} já existe.")
```
